```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(indian_movies['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from scipy.spatial.distance import cosine
```

```python
def recommendIndian(movie):
    movie_index = indian_movies[indian_movies['title'] == movie].index[0]
            
    similarity = cosine_similarity(vectors[movie_index].reshape(1,-1), vectors)
    distances = similarity[0]
    movies_list = sorted(list(enumerate(distances)), reverse = True, key = lambda x: x[1])[1:11]

    #     print(movies_list)
    for i in movies_list:
        print(indian_movies.loc[i[0]].title)
```
